```python
import pandas as pd

# Assuming you have a DataFrame called 'data' with your dataset
data = pd.get_dummies(full_data, columns=["Genre", "Director", "Actor 1", "Actor 2", "Actor 3"])
```

```python
# Label Encoding
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

# Assuming 'full_data' is your DataFrame
full_data["Director"] = label_encoder.fit_transform(full_data["Director"])
full_data["Genre"] = label_encoder.fit_transform(full_data["Genre"])
full_data["Actor 1"] = label_encoder.fit_transform(full_data["Actor 1"])
full_data["Actor 2"] = label_encoder.fit_transform(full_data["Actor 2"])
full_data["Actor 3"] = label_encoder.fit_transform(full_data["Actor 3"])
```

```python
full_data
```

```python
# remove rows
data.dropna(inplace=True)
```

```python
full_data
```

```python
# Fill with a Specific Value
data.fillna(0, inplace=True)  # Filling NaN values with 0
```

```python
full_data
```

```python
# Forward Fill or Backward Fill: If the missing values represent a time series or sequential data, you can use forward fill (propagate the last valid observation forward) or backward fill (propagate the next valid observation backward)
full_data.fillna(method='ffill', inplace=True)  # Forward fill
```

```python
# Interpolation: For time series data, interpolation methods like linear interpolation can be used to estimate missing values based on the surrounding data points.
data.interpolate(method='linear', inplace=True)
```

```python
full_data['Genre'].fillna('Unknown', inplace=True)
```
